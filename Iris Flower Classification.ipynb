```python

df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['Class'] = iris.target

df_iris
```

```python
def map_species(target):
    species_map = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
    return species_map[target]
```

```python
df_iris['Species'] = df_iris['Class'].apply(map_species)
```

```python
df_iris.drop('Class', axis=1, inplace=True)
```

```python

num_records, num_features = df_iris.shape
print(f"Number of records: {num_records}")
print(f"Number of features: {num_features}")
```

```python
records_per_class = df_iris['Species'].value_counts()
print("Number of records per class:")
print(records_per_class)
```

```python
missing_values = df_iris.isnull().sum()
print("Missing values in the DataFrame:")
print(missing_values)
```

```python
X = df_iris.drop('Species', axis=1)
y = df_iris['Species']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(train_X, train_y)
```

```python
predictions = knn_model.predict(test_X)
```

```python
accuracy = accuracy_score(test_y, predictions)
print("Model Accuracy:", accuracy)
```

```python
print("Classification Report:")
print(classification_report(test_y, predictions))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(test_y, predictions))
```
